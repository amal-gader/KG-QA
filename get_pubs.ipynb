```python
prefixes = """
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX orkgc: <http://orkg.org/orkg/class/>
PREFIX orkgp: <http://orkg.org/orkg/predicate/>
PREFIX orkgr: <http://orkg.org/orkg/resource/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>

"""

query = "SELECT DISTINCT ?URL\nWHERE {\n  orkgr:R112387 orkgp:compareContribution ?contrib.\n  ?contrib orkgp:HAS_DATASET ?URL.\n  FILTER(!REGEX(?URL, \"Not\"))\n}"
full_query = prefixes + query
```

```python
full_query="""
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX orkgc: <http://orkg.org/orkg/class/>
PREFIX orkgp: <http://orkg.org/orkg/predicate/>
PREFIX orkgr: <http://orkg.org/orkg/resource/>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>

SELECT ?paper ?link
WHERE {
  ?paper rdf:type orkgc:Paper .
  ?paper orkgp:P26 ?link .
}
LIMIT 100

"""
```

```python
import requests


def get_url(full_query):
    url = "http://localhost:7200/repositories/orkg_kg"
    headers = {
        "Accept": "application/sparql-results+json"
    }

    response = requests.get(url, params={'query': full_query}, headers=headers)
    urls = []

    if response.status_code == 200:
        data = response.json()
        for binding in data["results"]["bindings"]:
            print(f"https://www.doi.org/{binding['link']['value']}")
            urls.append(f"https://www.doi.org/{binding['link']['value']}")
    else:
        print(f"Error {response.status_code}: {response.text}")
    return urls
```

```python
urls = get_url(full_query)
```
