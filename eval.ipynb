```python
import os

def split_nt_file(input_path, chunk_size_gb=1):
    chunk_size_bytes = chunk_size_gb * 1024 * 1024 * 1024
    file_count = 1
    written_bytes = 0
    base_name, ext = os.path.splitext(input_path)
    output_file = open(f"{base_name}_chunk{file_count}{ext}", "w", encoding="utf-8")

    with open(input_path, "r", encoding="utf-8") as infile:
        for line in infile:
            line_size = len(line.encode("utf-8"))
            if written_bytes + line_size > chunk_size_bytes:
                output_file.close()
                file_count += 1
                output_file = open(f"{base_name}_chunk{file_count}{ext}", "w", encoding="utf-8")
                written_bytes = 0
            output_file.write(line)
            written_bytes += line_size

    output_file.close()
    print(f"âœ… Done! Created {file_count} chunks.")

# Example usage
split_nt_file("dblp_chunks/dblp.nt", chunk_size_gb=1)
```
